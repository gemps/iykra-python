```python
with open('tiga.json') as f:
    data = json.load(f)
```

```python
df = pd.json_normalize(data,'txs',[['user','id'],['user','name'],['user','email']],errors='ignore')
df
```

```python
num1 = df[['user.id','id']].set_index('user.id')
num1
```

```python
num2 = df[['user.id','user.name','user.email']].drop_duplicates().reset_index(drop=True).set_index('user.id')
num2
```
